```python
import numpy as np
import gzip#, cPickle
import _pickle as cPickle
from tsne import bh_sne
```

```python
output = np.load('train/output.npy').astype(np.float64)
data = np.load('train/data.npy')
target = np.load('train/target.npy')

print('data shape: ', data.shape)
print('target shape: ', target.shape)
print('output shape: ', output.shape)
```

```python
output_2d = bh_sne(output)
```

```python
np.save('train/output_2d.npy', output_2d, allow_pickle=False)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = 20, 20
plt.scatter(output_2d[:, 0], output_2d[:, 1], c=target)
plt.savefig('train/output_2d.png', bbox_inches='tight')
plt.show()
```

```python
data_num = data.shape[0]
data_new = np.reshape(data, (data_num, -1)).astype(np.float64)
print('data_new shape: ', data_new.shape)
```

```python
data_2d = bh_sne(data_new)
```

```python
data_2d.shape
```

```python
np.save('train/data_2d.npy', data_2d, allow_pickle=False)
```

```python
plt.rcParams['figure.figsize'] = 20, 20
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=target)
plt.savefig('train/data_2d.png', bbox_inches='tight')
plt.show()
```
